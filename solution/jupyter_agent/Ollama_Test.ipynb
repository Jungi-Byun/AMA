```python
import easyocr
import os

reader = easyocr.Reader(['ko', 'en'])  # 한국어+영어 지원

# 예시
exercises_dir = "./Exercises"
extracted_texts = []
for image_path in os.listdir(exercises_dir):
    if not image_path.endswith(('.jpg', '.png')):
        continue
    
    path = os.path.join(exercises_dir, image_path)
    results = reader.readtext(path, detail=0)
    problem_text = "\n".join(results)
    extracted_texts.append(problem_text)
    
    print(f"file: {path},\n {problem_text}")
    print("=" * 25)
```

```python
import requests
import json

def query_llama3(problem_text):
    # Ollama의 로컬 REST API 엔드포인트
    url = "http://localhost:11434/api/generate"

    # 프롬프트 작성
    prompt = f"""
        당신은 수학 문제를 친절하게 해설하고 풀어주는 초등학교 선생님입니다.
        최대 문장 생성 길이 내에서 이 문제의 해설 과정을 단계별로 자세히 알려주고, 마지막에 정답을 제시해주세요.
        초등학생도 이해할 수 있게 친절하고 간결하게 설명해주세요.

문제:
{problem_text}
"""

    payload = {
        "model": "llama3.1:8b",
        "prompt": prompt,
        "stream": False
    }

    headers = {
        "Content-Type": "application/json"
    }

    response = requests.post(url, headers=headers, data=json.dumps(payload))

    # JSON 결과
    result = response.json()
    return result["response"]
```

```python
for text in extracted_texts:
    print(query_llama3(text))
    print("=" * 25)
```
